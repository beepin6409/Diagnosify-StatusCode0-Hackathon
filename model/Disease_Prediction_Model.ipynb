```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Disease_with_Symptoms.csv")
df.head(20)
```

```python
df.shape
```

```python
#Since the values of the features are in the categorical format we have to do one hot encoding
data=pd.get_dummies(df.filter(regex='Symptom'),prefix='',prefix_sep='')
```

```python
data.head()
```

```python
data=data.groupby(data.columns,axis=1).agg(np.max)
```

```python
data.sample(20)
```

```python
filtered_data=data.iloc[:,1:-4]
```

```python
filtered_data.sample(10)
```

```python
filtered_data.shape
```

```python
Symptoms=filtered_data.columns
```

```python
filtered_data=pd.merge(filtered_data,df[['Code','Disease']],left_index=True,right_index=True)
```

```python
filtered_data.sample(10)
```

```python
disease_list=filtered_data[['Disease']]
```

```python
Disease_List=disease_list.drop_duplicates()
```

```python
disease=Disease_List.values.tolist()
```

```python
disease
```

```python
X=filtered_data.iloc[:,:-2]
Y=filtered_data.iloc[:,-2]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=5,test_size=0.2)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
Y_train.nunique()
```

```python
Y_test.nunique()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1=RandomForestClassifier(n_estimators=40)
model1.fit(X_train,Y_train)
```

```python
Y_predict=model1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y_test,Y_predict)*100
print(f"Accuracy of this model is {accuracy}%")
```

```python
def predict_disease(object):
    variable=model1.predict(object)[0]
    return disease[variable]
```
